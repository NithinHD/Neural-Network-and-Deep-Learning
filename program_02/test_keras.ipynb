```python
#using tensorflow
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.datasets import make_moons
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        
        # Initialize model
        self.model = Sequential()
        self.model.add(Dense(hidden_size, input_shape=(input_size,), activation='tanh'))
        self.model.add(Dense(output_size, activation='sigmoid'))

        # Compile the model
        self.model.compile(optimizer='adam', loss='binary_crossentropy')

    def train(self, X, Y, num_epochs, batch_size):
        self.model.fit(X, Y, epochs=num_epochs, batch_size=batch_size, verbose=1)

    def predict(self, X):
        return self.model.predict(X)
```

```python

# Load and preprocess the data
X, Y = make_moons(n_samples=1000, noise=0.2, random_state=42)
X = X.T.astype(np.float32).T
Y = Y.reshape(1, -1).astype(np.float32).T
print(X.shape, Y.shape)
```
